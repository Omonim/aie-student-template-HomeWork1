```python
import numpy as np

# массивы
arr1 = np.array([1, 2, 3, 4, 5])
arr2 = np.array([[1, 2, 3], [4, 5, 6]])
print("arr1:", arr1)
print("arr2:")
print(arr2)
# нули и единицы
zeros_arr = np.zeros(5)
ones_arr = np.ones((2, 3))
print("zeros:", zeros_arr)
print("ones:")
print(ones_arr)
# arange
range_arr = np.arange(0, 10, 2)
print("arange:", range_arr)
# атрибуты
print("shape arr2:", arr2.shape)
print("dtype arr2:", arr2.dtype)
# индексы и срезы
print("arr2[1,2]:", arr2[1, 2])
print("arr1[1:4]:", arr1[1:4])
print("arr2 срез:")
print(arr2[0:2, 1:3])
# векторные операции
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
print("a + b:", a + b)
print("a * b:", a * b)
# статистика
print("mean a:", np.mean(a))
print("sum b:", np.sum(b))
print("max a:", np.max(a))
print("min b:", np.min(b))
print("std a:", np.std(a))
# статистика для матрицы
matrix = np.array([[1, 2], [3, 4]])
print("matrix mean:", np.mean(matrix))
print("mean axis=0:", np.mean(matrix, axis=0))
print("mean axis=1:", np.mean(matrix, axis=1))
```
